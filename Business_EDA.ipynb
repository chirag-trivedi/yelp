**Info about business**

```
import pandas as pd

df_business = pd.read_json(r'/content/drive/My Drive/yelp/yelp_academic_dataset_business.json', lines=True)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
print(df_business.head())
df_business.info()
df_business.shape
```

**States with most businesses**

```
df_count_business_id = df_business.groupby('state').count()[['business_id']]
df_count_business_id.sort_values(by='business_id', ascending=False)
```

**Businesses with stars 3 and above**

```
df_business.query('stars >= 3')
```

**Businesses sorted by reviews**

```
df_business.sort_values(by='review_count', ascending=False)[['name','city','state','stars','review_count','attributes','categories']]
```

**Business Count by City in Arizona**

**Business Count in Phoenix, Arizona**

Taking care of misspellings and case


```
display(df_business[df_business['state']=='NV']['city'].str.lower().value_counts().filter(like = 'veg'))
df_business[df_business['state']=='NV']['city'].str.lower().str.contains("vegas").value_counts()
```

**Categories**

```
category_count = {}

for item in df_business['categories']:
  if item is None:
    category_list = ['N/A']
  else:  
    category_list = item.split(',')
  for category in category_list:
    try:
      category_count[category] += 1
    except:
      category_count[category] = 1  
print(category_count)
       
```
